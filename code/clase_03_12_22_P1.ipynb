### Vamos a volver a cargar el archivo en la misma variable, pero reemplazaremos los nombres de las columnas, vamos a reemplazar el tipo de dato en month y en year.

```python
temp = pd.read_csv('data/landtempssample.csv', names=['estaciónID','año', 'mes', 
                                                      'promedioT', 'latitud', 
                                                      'longitud', 'elevación',
                                                     'estación', 'paísID', 'país'], skiprows=1, 
                   parse_dates=[['año','mes']])
```

```python
temp.head(10)
```

```python
temp.dtypes
```

```python
temp.shape
```

```python
temp['día']=15
```

```python
temp.head(10)
```

```python
temp.rename(columns={'año_mes':'fecha_muestreo'})
```

```python
temp
```

Al ejecutar rename, solo se cambia el nombre de la columna en esa ejecución del código, para realizar el cambio de forma permanente debes usar `inplace=True`

```python
temp.rename(columns={'año_mes':'fecha_muestreo'},inplace=True)
```
