
### Vamos a utilizar una base de datos pública que tiene datos de educación

````python
# Definiremos una query para traer los datos

query = "SELECT studentid, school, sex, age, famsize, medu AS mothereducation, \
fedu AS fathereducation, traveltime, studytime,failures, famrel, freetime, goout,\
g1 AS gradeperiod1, g2 AS gradeperiod2, g3 AS gradeperiod3 FROM studentmath"
````

````python
#Ahora guardaremos las credenciales de acceso en variables:
server = 'pdcc.c9sqqzd5fulv.us-west-2.rds.amazonaws.com'
user= "pdccuser"
password="pdccpass"
database = "pdcctest"
````

````python
# Usaremos la api de pymssql para traernos la data del Server SQL

conexion=pymssql.connect(server=server, user=user, password=password, database=database)
````

````python
matematica=pd.read_sql(query,conexion)
conexion.close()
````
