Como bien dijo Fernando, no encontramos una columna `birth_year`, sin embargo, podemos agregar esta columna, procesando (como se llama este curso), los datos dentro de nuestro dataframe.

Vamos a crear una función `lambda`:  
```python

conjuntoDatosDF.creators.apply(lambda x: x[0]['birth_year'])

```
Del `conjuntoDatosDF` estamos seleccionando la serie `creators`,
donde `x` representa a la serie (columna) `creators`, entonces `x[0]` nos dará `creators[0]` y extraeremos desde ahí la `llave` `birth_year`

````python
conjuntoDatosDF['birth_year'] = conjuntoDatosDF.creators.apply(lambda x: x[0]['birth_year']) 
````
